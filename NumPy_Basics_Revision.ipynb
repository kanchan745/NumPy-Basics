```python
import numpy as np
#speed high numpy and take less memory ,easy math operation ,large amount of data
#numpy array is a special type of list
temperatures = np.array([32.5,85.4,56.2])
average = np.mean(temperatures)
print(average)
```

```python
#array : collection of no./values which can be store together in an organize manner
#tpes 1d ,2d ,3d ,multidimensional _array

numpy_array = [1,2,4,5]
print(numpy_array)
```

**1 Dimensional array**

one dimensional array it is simply a list of  no. that stores in a single row

```python
ar_1d = np.array([10,20,30,40,50])
print(ar_1d)
```

**2 Dimensional array**

Two dimensional array : rows + columns

```python
arr_2d = np.array([[1,2,3],[4,5,6],[5,5,8]])
print(arr_2d)
```

**Multi-dimensional array**

A array that have multi dimension means 2d ,3d ,4d... use in Ai, Dl,MRI SCAN, 3D MODELLING, or we can say it has multiple layer of storage data, matrix-2d array

```python
matrix = np.array([[2,4,6],[8,10,12]])
print(matrix)
```

# Creating arrays from python lists

**2nd Method with Default Values ....**

for which numpy provide built in function which is 0's function : use for future values
syntax -> np.zeros(shape)

```python
zeroes_array = np.zeros(3)
print(zeroes_array)
```

**for 1's values** ones (shape)

```python
ones_array = np.ones((2,3))#tuple form
print(ones_array)
```

**For specific values : full function** synatax: (shapes,value)

```python
filled_array = np.full((2,2),7)
print(filled_array)
```

# Sequences of numbers in numpy function : arange function (), syntax : arange(start,stop,step)

```python
arr = np.arange(1,10,2)
print(arr)
```
